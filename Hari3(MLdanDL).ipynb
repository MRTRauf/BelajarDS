# DATA **IRIS**

```
import keras
import numpy as np
import pandas as pd
import seaborn as sns
from keras import Sequential
from keras.layers import Dense
```

```
dataset = pd.read_csv('https://gist.githubusercontent.com/netj/8836201/raw/6f9306ad21398ea43cba4f7d537619d0e07d5ae3/iris.csv')
dataset
```

```
#PISAH INPTUT OUTPUT
x = dataset.iloc[:,:-1].values
y = pd.get_dummies(dataset["variety"]).values
```

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 42)   #test_size = persenan data test, sisanya data train
```

```
X_train
```

```
x.shape
```

```
classifier = Sequential()

# INISIASI INPUT LAYER SEBANYAK 2 LAYER DAN HIDDEN LAYER SEBANYAK 3 NODE
classifier.add(Dense(units = 50, kernel_initializer = 'uniform', activation = 'relu', input_dim = 4))

# PENDAMBAHAN HIDDEN LAYER KEDUA 
classifier.add(Dense(units = 30, kernel_initializer = 'uniform', activation = 'relu'))
# PENDAMBAHAN HIDDEN LAYER KEDUA 
classifier.add(Dense(units = 30, kernel_initializer = 'uniform', activation = 'relu'))

# PENAMBAHAN OUTPUT LAYER
classifier.add(Dense(units = 3, kernel_initializer = 'uniform', activation = 'softmax'))  #SIGMOID MISAL INPUT 1/0

# COMPILE ANN
classifier.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])   #bergantung pada output (kalau 1 dan 0 maka binary, )
```

```
classifier.fit(X_train, y_train, batch_size = 10, epochs = 60, validation_data=(X_test,y_test))
```

```
hasil = classifier.predict(X_test)
```

```
hasil
```

```
hasil = np.round(hasil)
test = np.array(y_test)
validasi = pd.DataFrame()
validasi["Setosa"] = test.T[0]
validasi["Versicolor"] = test.T[1]
validasi["Virginica"] = test.T[2]
validasi["Pred Setosa"] = hasil.T[0]
validasi["Pred Versicolor"] = hasil.T[1]
validasi["Pred Virginica"] = hasil.T[2]
```
